# float()

### arithemetic Operations (+, -, *, /)

```python
# 1. Calculating average score precisely
total_score = 485.0 
num_students = 50
average_score = total_score / num_students
print(f"Precise average score: {average_score}")
```

```python

# 2. Calculating completion percentage
completed_modules = 7.0
total_modules = 15
completion_rate = (completed_modules / total_modules) * 100
print(f"Course completion rate: {completion_rate:.2f}%")
```

```python
# 2. Calculating completion percentage
completed_modules = 7.0
total_modules = 15
completion_rate = (completed_modules / total_modules) * 100
print(f"Course completion rate: {completion_rate:.f}%")
```

```python
# 3. Dividing integers results in float
revenue = 750000
total_students = 1500
revenue_per_student = revenue / total_students
print(f"Revenue per student: {revenue_per_student}")
```

```python
# 4. Simple division
val1 = 10.0
val2 = 4.0
result = val1 / val2
print(f"10.0 / 4.0 = {result}")
```

```python
# 5. Division involving zero
result = 0.0 / 5.0
print(f" 0.0 / 5.0 = {result}")
```

```python
# 6. Division by small number
value = 1.0
divisor = 0.001
result = value / divisor
print(f"1.0 / 0.001 = {result}")
```

```python
# 7. Using division in scaling
original_rating = 4.8
scale_factor = 2.0 # Example: Adjusting scale
scaled_rating = original_rating / scale_factor
print(f"Scaled rating: {scaled_rating}")
```

```python
# 8. Negative division
val1 = -15.0
val2 = 2.0
result = val1 / val2
print(f"-15.0 / 2.0 = {result}")
```

```python
# 8. Negative division
val1 = -15.0
val2 = 2.0
result = abs(val1 / val2)    ## using abs() to get the positive value of the result
print(f"-15.0 / 2.0 = {result}")
```

```python
# 9. Division resulting in repeating decimal (shows precision limits)
result = 1.0 / 3.0
print(f"1.0 / 3.0 = {result}")
```

```python
# 9. Division resulting in repeating decimal (shows precision limits)
result = 1.0 / 3.0
print(f"1.0 / 3.0 = {result:.3f}")  ## using :.3f to format the output to 3 decimal places
```

```python
# 10. Calculating average monthly growth
total_growth = 55.5 # e.g. average increase over period
months = 6
avg_monthly = total_growth / months
print(f"Average monthly growth: {avg_monthly}")
```

```python
# 10. Calculating average monthly growth
total_growth = 125.5
months = 6
avg_monthly = total_growth / months
print(f"Average monthly growth: {avg_monthly:.2f}")  ## using :.2f to format the output to 2 decimal places
```

### Floor division (//) adn Modulo (%)

```python
# 1. How many full 'rating points' in average_rating
average_rating = 4.8
full_points = average_rating // 1.0
print(f"Full rating points in {average_rating}: {full_points}")
```

```python
# 2. Number of full $10 units in a price
price = 49.95
units_of_10 = price // 10.0
print(f"Full $10 units in {price}: {units_of_10}")
```

```python
# 3. Dividing floats with floor division
val1 = 10.5
val2 = 2.1
result = val1 // val2
print(f"10.5 // 2.1 = {result}")
```

```python
# 4. Negative float floor division
val1 = -10.5
val2 = 2.1
result = val1 // val2
print(f"-10.5 // 2.1 = {result}")
```

```python
# 4. Negative float floor division
val1 = -10.5
val2 = 2.1
result = abs(val1 // val2) ## using abs() to get the positive value of the result
print(f"-10.5 // 2.1 = {result}")
```

```python
# 5. Using very small divisor
val1 = 1.0
val2 = 0.3
result = val1 // val2
print(f"1.0 // 0.3 = {result}")
```

```python
# 6. Floor division by 1.0
val = 7.8
result = val // 1.0
print(f"7.8 // 1.0 = {result}")
```

```python
# 7. Floor division result is zero
val1 = 2.5
val2 = 3.0
result = val1 // val2
print(f"2.5 // 3.0 = {result}")
```

```python
# 8. Calculating full weeks from fractional days
days = 30.5
days_per_week = 7.0
full_weeks = days // days_per_week
print(f"Full weeks in {days} days: {full_weeks}")
```

```python
# 9. Large numbers
val1 = 1000.0
val2 = 150.5
result = val1 // val2
print(f"1000.0 // 150.5 = {result}")
```

```python
# 10. Floor division involving integers (promotes to float)
result = 10 // 3.0
print(f"10 // 3.0 = {result}")
```

#### round(number[,ndigits])

```python
# 1. Rounding the average rating to one decimal place
average_rating = 4.8356
rounded_rating_1dp = round(average_rating, 1)
print(f"Average rating rounded to 1dp: {rounded_rating_1dp}")
```

```python
# 1. Rounding the average rating to one decimal place
average_rating = 4.8356
rounded_rating_1dp = round(average_rating, 3)  ## using 3 to round to 3 decimal places          
print(f"Average rating rounded to 3dp: {rounded_rating_1dp}")
```

```python
# 2. Rounding the average rating to the nearest integer
average_rating = 4.8356
rounded_rating_int = round(average_rating)
print(f"Average rating rounded to nearest int: {rounded_rating_int}")
```

```python
# 3. Rounding completion rate to two decimal places
completion_rate = 46.6666666
rounded_rate_2dp = round(completion_rate, 2)
print(f"Completion rate rounded to 2dp: {rounded_rate_2dp}")
```

```python
# 4. Rounding to zero decimal places (returns float)
price = 49.95
rounded_price_0dp = round(price, 0)
print(f"Price rounded to 0dp: {rounded_price_0dp}")
```

```python
# 5. Round half to even: 2.5
result = round(2.5)
print(f"round(2.5): {result}")
```

```python
# 6. Round half to even: 3.5
result = round(3.5)
print(f"round(3.5): {result}")
```

```python
# 7. Rounding a negative number
neg_value = -4.8356
rounded_neg_1dp = round(neg_value, 1)
print(f"-4.8356 rounded to 1dp: {rounded_neg_1dp}")
```

```python
# 8. Rounding to nearest integer (negative)
rounded_neg_int = round(neg_value)
print(f"-4.8356 rounded to nearest int: {rounded_neg_int}")
```

```python
# 8. Rounding to nearest integer (negative)
rounded_neg_int = round(abs(neg_value)) ## using abs() to get the positive value of the result
print(f"-4.8356 rounded to nearest int: {rounded_neg_int}")
```

```python
# 9. Rounding to negative digits (rounds to tens, hundreds...)
total_students_precise = 1537.5
rounded_to_tens = round(total_students_precise, -1)
print(f"Students rounded to nearest 10: {rounded_to_tens}")
```

```python
# 10. Rounding to nearest hundred
revenue = 753450.75
rounded_revenue_hundreds = round(revenue, -2)
print(f"Revenue rounded to nearest 100: {rounded_revenue_hundreds}")
```

```python
# 10. Rounding to nearest hundred
revenue = 753450.75 
rounded_revenue_hundreds = round(revenue, 0)  ## using 0 to round to nearest 10 instead of 100
print(f"Revenue rounded to nearest 100: {rounded_revenue_hundreds}")
```

#### float.is_integer()

```python
# 1. Check if average rating is an integer
average_rating = 4.8
print(f"Is {average_rating} an integer? {average_rating.is_integer()}")
```

```python
average_rating = 4
print(f"Is {average_rating} an integer? {average_rating.is_integer()}")
```

```python
# 2. Check a float that is an integer
revenue_per_student = 500.0
print(f"Is {revenue_per_student} an integer? {revenue_per_student.is_integer()}")
```

```python
# 3. Check zero
zero_float = 0.0
print(f"Is {zero_float} an integer? {zero_float.is_integer()}")
```

```python
# 4. Check a negative integer float
neg_int_float = -10.0
print(f"Is {neg_int_float} an integer? {neg_int_float.is_integer()}")
```

```python
# 5. Check a negative non-integer float
neg_non_int_float = -10.5
print(f"Is {neg_non_int_float} an integer? {neg_non_int_float.is_integer()}")
```

```python
# 6. Result of a calculation
result = 10.0 / 2.0 # 5.0
print(f"Is result of 10.0/2.0 an integer? {result.is_integer()}")
```

```python
# 7. Result of another calculation
result = 10.0 / 3.0 # 3.333...
print(f"Is result of 10.0/3.0 an integer? {result.is_integer()}")
```

```python
# 8. Float representation of total students
float_students = float(1500)
print(f"Is float(1500) an integer? {float_students.is_integer()}")
```

```python
# 9. Very small fractional part (may be False due to precision)
small_frac = 1.0000000000000001
print(f"Is {small_frac} an integer? {small_frac.is_integer()}")
```

```python
# 10. A float resulting from rounding
rounded_val = round(4.8) # Becomes int 5, but if result was float like round(4.8, 0) -> 5.0
rounded_float = round(4.95, 0) # 5.0
print(f"Is round(4.95, 0) an integer? {rounded_float.is_integer()}")
```

### float.as_integer_ratio()

```python
# 1. Ratio for 0.5
ratio = (0.5).as_integer_ratio()
print(f"Ratio for 0.5: {ratio}")
```

```python
# 2. Ratio for average rating (shows potential complexity)
average_rating = 4.8
ratio = average_rating.as_integer_ratio()  # 4.8 = 48/10 = 24/5
print(f"Ratio for {average_rating}: {ratio}")

ratio_simple = (4.75).as_integer_ratio() # 4.75 = 19/4
print(f"Ratio for 4.75: {ratio_simple}")
```

```python
# 3. Ratio for an integer float
ratio = (500.0).as_integer_ratio()
print(f"Ratio for 500.0: {ratio}")
```

```python
# 4. Ratio for 0.25
ratio = (0.25).as_integer_ratio()
print(f"Ratio for 0.25: {ratio}")
```

```python
# 5. Ratio for a negative float
ratio = (-0.75).as_integer_ratio()
print(f"Ratio for -0.75: {ratio}")
```

```python
# 6. Ratio for zero
ratio = (0.0).as_integer_ratio()
print(f"Ratio for 0.0: {ratio}")
```

```python
# 7. Ratio demonstrating precision limits (0.1 is not exact in binary)
ratio = (0.1).as_integer_ratio()
print(f"Ratio for 0.1 (precision): {ratio}")
```

```python
# 8. Ratio for 1.5
ratio = (1.5).as_integer_ratio()
print(f"Ratio for 1.5: {ratio}")
```

```python
# 9. Ratio for a large float
large_float = 12345.6789
ratio = large_float.as_integer_ratio()
print(f"Ratio for {large_float} (can be large): {ratio}")
```

```python
# 10. Ratio from a calculation result
result = 10.0 / 8.0 # 1.25
ratio = result.as_integer_ratio()
print(f"Ratio for 1.25: {ratio}")
```

#### float()

```python
# 1. Converting total students (int) to float
total_students = 1500
students_float = float(total_students)
print(f"Total students as float: {students_float}")
```

```python
# 2. Converting integer 0 to float
zero_float = float(0)
print(f"Integer 0 as float: {zero_float}")
```

```python
# 3. Converting a string representation of a float
average_rating_str = "4.8"
average_rating_float = float(average_rating_str)
print(f"Average rating from string: {average_rating_float}")
```

```python
# 4. Converting a string representation of an integer
int_str = "100"
int_str_float = float(int_str)
print(f"Integer string '100' as float: {int_str_float}")
```

```python
# 5. Converting a string with scientific notation
sci_str = "1.5e3" # 1.5 * 10^3
sci_float = float(sci_str)
print(f"Float from '1.5e3': {sci_float}")
```

```python
# 6. Converting boolean True to float
true_float = float(True)
print(f"Float value of True: {true_float}")
```

```python
# 7. Converting boolean False to float
false_float = float(False)
print(f"Float value of False: {false_float}")
```

```python
# 8. Converting special strings 'inf', '-inf', 'nan'
inf_float = float('inf')
print(f"Float from 'inf': {inf_float}") # Output: inf
neg_inf_float = float('-inf')
print(f"Float from '-inf': {neg_inf_float}") # Output: -inf
nan_float = float('nan')
print(f"Float from 'nan': {nan_float}")
```

```python
# 9. Converting a negative integer string
neg_int_str = "-50"
neg_int_float = float(neg_int_str)
print(f"Float from '-50': {neg_int_float}")
```

```python
# 9. Converting a negative integer string
neg_int_str = "-50"
neg_int_float = float(abs(int(neg_int_str)))  ## using abs() to convert the negative integer string to positive before converting to float
print(f"Float from '-50': {neg_int_float}")
```

```python
# 10. Attempting invalid conversion (raises ValueError)
try:
    invalid_float = float("Edukron 4.8")
except ValueError as e:
    print(f"Error converting 'Edukron 4.8' to float: {e}")
```
