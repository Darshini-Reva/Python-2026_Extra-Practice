#### Floor Division (//)

```python
# 1. Calculating number of full batches
total_students = 157
students_per_batch = 50

full_batches = total_students // students_per_batch
print(full_batches)
print(f"Number of full batches: {full_batches}")
```

```python
# 3. Calculating completed weeks
days_elapsed = 30
days_per_week = 7

completed_weeks = days_elapsed // days_per_week
print(f"Completed weeks: {completed_weeks}")
```

```python
# 4. Dividing by a larger number
small_group = 10
large_capacity = 100

num = small_group // large_capacity
print(f"Large number: {num}")
```

```python
# 6. Floor division with negative numbers (rounds down)
debt = -100
payments_per_month = 30

months = debt // payments_per_month
print(f"Payemnt for no of months: {months}")
```

```python
# 7. Another negative example
val1 = -10
val2 = 3

val = val1 // val2
print(val)
```

```python
# 8. Positive numerator, negative denominator
val1 = 10
val2 = -3

val = val1 // val2
print(val)
```

```python
# 9. Calculating full sets of resources
total_licenses = 125
licenses_per_set = 10

full_set = total_licenses // licenses_per_set
print(f" Full set of resources: {full_set}")
```

```python
# 10. Integer part of an average (if calculated stepwise)
total_score = 485
num_students = 50

part_average = total_score // num_students
print(f"Integer part of an average: {part_average}")
```

#### Exponentiation (**)

```python
# 1. Modeling simplified exponential student growth (base * factor^periods)
initial_students = 100
growth_factor = 2
periods = 3

students = initial_students * (growth_factor ** periods)
print(students)
```

```python
# 2. Calculating squares (power of 2)
side_length = 5
area = side_length ** 2

print(f"Area of square: {area}")
```

```python
# 3. Calculating cubes (power of 3)
edge_length = 3
volume = edge_length ** 3

print(volume)
```

```python
# 4. Power of zero (always 1, except 0**0 is debated but 1 in Python)
base = 10
result = base ** 0
print(f" 10 to the power of 0: {result}")

zero_base = 0
result_zero = zero_base ** 0
print(f" 0 to the power of 0: {result_zero}")
```

```python
# 5. Power of one (always the base itself)

total_students = 1500
base = total_students
result = base ** 1

print(result)
```

```python
# 6. Large powers (possible due to arbitrary precision)
large_power = 2 ** 100
print(large_power)
print(type(large_power))
print(str(large_power)[:10])
```

```python
# 7. Negative base, even power (positive result)
base = -5
power = 2
result = base ** power

print(result)
```

```python
# 8. Negative base, odd power (negative result)
base = -5
power = 3
result = base ** power

print(result)
```

```python
# 9. Using exponentiation for bit shifting (2**n is like 1 << n)
bit_shift_equivalent = 2 ** 3

print(bit_shift_equivalent)
```

```python
# 10. Calculating compound interest (simplified, int only)
principal = 10000
rate_percent = 5 # Integer percent
periods = 3

approx_value = principal * (100 + rate_percent) ** periods // (100 ** periods)
print(approx_value)
```

### abs ()

```python
# 1. Getting absolute change in students

student_change = -50
students = abs(student_change)
print(students)
```

```python
# 2. Absolute value of a positive number
positive_count = 1500
abs_count = abs(positive_count)

print(abs_count)
```

```python
# 3. Absolute value of zero
zero_value = 0
abs_zero = abs(zero_value)

print(abs_zero)
```

```python
# 4. Ensuring a difference is positive
score1 = 85
score2 = 92

score = abs(score1 - score2)
print(score)
```

```python
# 5. Absolute error calculation
estimated_students = 1450
actual_students = 1500

error = abs(estimated_students - actual_students)
print(error)
```

```python
# 6. Using abs() in calculations
value = -10
result = abs(value) * 5
print(result)
```

```python
# 7. Absolute value from a calculation result
net_profit = 10000 - 15000 # (-5000)
abs_profit_or_loss = abs(net_profit)
print(abs_profit_or_loss)
```

```python
# 8. Applying to batch id difference
batch_id1 = 101
batch_id2 = 105
id_distance = abs(batch_id1 - batch_id2)
print(id_distance)
```

```python
# 9. Absolute value of a large negative number
large_neg = -999999999
abs_large = abs(large_neg)
print(abs_large)
print(str(abs_large)[:4])
```

```python
# 10. Absolute value within a conditional check (conceptual)
deviation = -5
if abs(deviation) > 3:
    print(deviation)
```

### pow (base, exp[,mod])

```python
# 1. Simple power calculation
result = pow(10, 3)

print(result)
```

```python
# 2. Equivalent to growth example above
initial_students = 100
growth_factor = 2
periods = 3

projected_students = initial_students * pow(growth_factor,periods)
print(projected_students)
```

```python
# 3. Power of 2
side_length = 5 
area = pow(side_length, 2)
print(area)
```

```python
# 4. Power of 0

total_students = 1500
base = total_students 
result = pow(base, 0)
print(result)
```

```python
# 5. Power of 1
base = 42
result = pow(base, 1)
print(result)
```

```python
# 6. Negative base, even power
result = pow(-5, 2)
print(result)
```

```python
# 7. Negative base, odd power
result = pow(-5, 3)
print(result)
```

```python
# 8. Modular exponentiation: (3 ** 4) % 5
result_mod = pow(3, 4, 5)
print(result_mod)
```

```python
# 9. Modular exponentiation: (2 ** 10) % 7
result_mod_large = pow(2, 10, 7)
print(result_mod_large)
```

```python
# 10. Modular exponentiation with negative base (handle carefully based on definition)
# pow(-2, 3, 5) -> (-2)^3 = -8. (-8) % 5 -> Python gives 2 (-8 = -2*5 + 2)

result_mod_neg = pow(-2, 3, 5)
print(result_mod_neg)
```

### divmod(a,b)

```python
# 1. Getting full batches and remaining students simultaneously
total_students = 157
students_per_batch = 50

batches, remaining = divmod(total_students, students_per_batch)
print(f"batches :{batches}, remaining : {remaining}")
```

```python
# 2. Distributing assignments and finding leftovers
total_assignments = 47
num_instructors = 4
assign_per, leftover_assign = divmod(total_assignments, num_instructors)
print(f" Assignments per instructor: {assign_per}, Leftovers: {leftover_assign}")
```

```python
# 3. Calculating weeks and remaining days
days_elapsed = 30
days_per_week = 7
weeks, rem_days = divmod(days_elapsed, days_per_week)
print(f"Weeks: {weeks}, Remaining Days: {rem_days}")
```

```python
days_elapsed = 30   ## tried in simplier version 
days_per_week = 7

weeks = divmod(days_elapsed,days_per_week)
print(weeks)
```

```python
# 4. Dividing by a larger number
small_group = 10
large_capacity = 100
quotient, remainder = divmod(small_group, large_capacity)
print(f"divmod(10, 100): Quotient={quotient}, Remainder={remainder}") 
```

```python
# 5. Dividing zero
zero_students = 0
batch_size = 50
quotient, remainder = divmod(zero_students, batch_size)
print(f"divmod(0, 50): Quotient={quotient}, Remainder={remainder}")
```

```python
# 6. Negative numerator
debt = -100
payments_per_month = 30
q, r = divmod(debt, payments_per_month)
print(f"divmod(-100, 30): Quotient={q}, Remainder={r}")
```

```python
# 7. Another negative numerator example
q, r = divmod(-10, 3)
print(f"divmod(-10, 3): Quotient={q}, Remainder={r}")
```

```python
# 8. Positive numerator, negative denominator
q, r = divmod(10, -3)
print(f"divmod(10, -3): Quotient={q}, Remainder={r}")
```

```python
# 9. Calculating full sets and leftovers
total_licenses = 125
licenses_per_set = 10
sets, leftover_licenses = divmod(total_licenses, licenses_per_set)
print(f"Full sets: {sets}, Leftover licenses: {leftover_licenses}")
```

```python
# 10. Converting total minutes to hours and minutes
total_minutes = 135
minutes_per_hour = 60
hours, minutes = divmod(total_minutes, minutes_per_hour)
print(f"{total_minutes} mins = {hours} hours and {minutes} minutes")
```

#### int ()

```python
# 1. Converting average rating (float) to int (truncates)
average_rating = 4.8
rating_int = int(average_rating)
print(f"Average rating as int: {rating_int}")
```

```python
# 2. Converting a string representation of an integer
total_students_str = "1500"
print(type(total_students_str))
total_students_int = int(total_students_str)
print(f"Total students from string: {total_students_int}")
```

```python
# 3. Converting a negative float string
neg_float_str = "-10.5"
neg_int = int(float(neg_float_str)) # Need float() first if string has decimal
print(f"Integer from '-10.5': {neg_int}")
```

```python
# 4. Converting boolean True to int
is_active_int = int(True)
print(f"Integer value of True: {is_active_int}")
```

```python
# 5. Converting boolean False to int
is_full_int = int(False)
print(f" Integer value of False: {is_full_int}")
```

```python
# 6. Converting a positive float
positive_float = 99.99
positive_int = int(positive_float)
print(f"Integer from 99.99: {positive_int}")
```

```python
# 7. Converting a negative float
negative_float = -0.1
negative_int = int(negative_float)
print(f" Integer from -0.1: {negative_int}")
```

```python
# 8. Converting string with base (e.g., binary)
binary_str = "1101" # Represents 13
decimal_val = int(binary_str, 2)
print(f"Integer from binary '1101': {decimal_val}")
```

```python
# 8. Converting string with base 
binary_str = "101"
decimal_val = int(binary_str, 3)
print(f"Integer from binary '101': {decimal_val}")
```

```python
# 9. Converting string with base (e.g., hexadecimal)
hex_str = "FF" # Represents 255
decimal_val_hex = int(hex_str, 16)
print(f"Integer from hex 'FF': {decimal_val_hex}")
```

```python
# 9. Converting string with base 
hex_str = "EE" 
decimal_val_hex = int(hex_str, 16)
print(f"Integer from hex 'EE': {decimal_val_hex}")
```
